```python
import sqlite3
import pandas as pd
```

```python
def get_table(table_name):
    conn = sqlite3.connect('Sql_board.sqlite')
    conn.row_factory = sqlite3.Row
    db = conn.cursor()

    rows = db.execute(f"SELECT * FROM {table_name}").fetchall()

    conn.commit()
    conn.close()

    return [dict(row) for row in rows]
```

```python
# example
pd.DataFrame(get_table('boardprice'))
```

```python
get_table('bbg')
```
