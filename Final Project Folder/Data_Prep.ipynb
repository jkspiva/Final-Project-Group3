```python
import pandas as pd

# Adjust the delimiter based on your file's structure
price_df = pd.read_csv('./Resources/board_game_price.csv', delimiter=',')

# Display the first few rows
print(price_df.head())
```

```python
# Calculate the percentage of users rated and round to 2 decimal places
price_df['pct_rated'] = round(price_df['Users Rated'] / price_df['Owned Users'], 2)

price_df['pct_rated'].replace([float('inf'), float('-inf')], 0, inplace=True)

# Display the first few rows of the DataFrame
print(price_df.head())
```

```python
price_df['Domain1'] = price_df['Domain1'].fillna('other')
```

```python
# Replace NaN values with 0 in the 'Bayesian Average' column
price_df['Bayesian Average'].fillna(0, inplace=True)

# Verify the replacement by checking if there are any remaining NaNs
null_count_after = price_df['Bayesian Average'].isna().sum()

print(f'Number of null values after replacement: {null_count_after}')
```

```python
# Strip any leading/trailing spaces or hidden characters
price_df['Bayesian Average'] = price_df['Bayesian Average'].str.strip()

# Replace any remaining non-numeric values with NaN, then fill NaNs with 0
price_df['Bayesian Average'] = pd.to_numeric(price_df['Bayesian Average'], errors='coerce').fillna(0)

# Round the values to 2 decimal places
price_df['Bayesian Average'] = price_df['Bayesian Average'].round(2)

# Verify the conversion
print(price_df['Bayesian Average'].dtype)
```

```python
# Round the values in the 'Bayesian Average' column to 2 decimal places
price_df['Bayesian Average'] = price_df['Bayesian Average'].round(2)
```

```python
# Calculate the range (min and max)
min_value = price_df['Bayesian Average'][price_df['Bayesian Average'] > 0].min()
max_value = price_df['Bayesian Average'].max()
value_range = max_value - min_value

# Calculate the median
median_value = price_df['Bayesian Average'].median()

# Display the results
print(f"Minimum Value: {min_value}")
print(f"Maximum Value: {max_value}")
print(f"Range: {value_range}")
print(f"Median Value: {median_value}")
```

```python
price_df['bayes_above_average'] = price_df['Bayesian Average'] > 5.54
price_df.head()
```

```python
price_df['Owned Users'].fillna(0, inplace=True)
# Calculate the mean of the non-zero 'Price' values
non_zero_mean = price_df[price_df['Owned Users'] > 0]['Owned Users'].mean()

# Print the calculated mean
print(f'The mean of non-zero Owners is: {non_zero_mean}')

price_df.loc[price_df['Owned Users'] == 0, 'Owned Users'] = non_zero_mean
```

```python
price_df[['Most Notable Genre', 'Most Notable Mechanic']] = price_df[['Most Notable Genre', 'Most Notable Mechanic']].fillna('unspecified')
```

```python
# Replace NaN values with 0 in the 'Bayesian Average' column
price_df['Price'].fillna(0, inplace=True)

# Verify the replacement by checking if there are any remaining NaNs
null_count_after = price_df['Price'].isna().sum()

print(f'Number of null values after replacement: {null_count_after}')
```

```python
# Strip any leading/trailing spaces or hidden characters
price_df['Price'] = price_df['Price'].str.strip()

# Replace any remaining non-numeric values with NaN, then fill NaNs with 0
price_df['Price'] = pd.to_numeric(price_df['Price'], errors='coerce').fillna(0)

# Round the values to 2 decimal places
price_df['Price'] = price_df['Price'].round(2)

# Verify the conversion
print(price_df['Price'].dtype)

price_df['Price'] = price_df['Price'].round(2)
```

```python
# Calculate the range (min and max)
min_value = price_df['Price'][price_df['Price'] > 0].min()
max_value = price_df['Price'].max()
value_range = max_value - min_value

# Calculate the median
median_value = price_df['Price'].median()

# Display the results
print(f"Minimum Value: {min_value}")
print(f"Maximum Value: {max_value}")
print(f"Range: {value_range}")
print(f"Median Value: {median_value}")
```

```python
import pandas as pd

# Adjust the delimiter based on your file's structure
df2 = pd.read_csv('./Resources/bgg_dataset.csv', delimiter=';')

# Display the first few rows
print(df2.head())
```

```python
import pandas as pd

# Perform the merge
df_merged = pd.merge(df, df2[['Name', 'Year Published']], 
                     on='Name', 
                     how='left')

df_merged.head()
```

```python
null_counts = df_merged.isnull().sum()
print(null_counts)
```

```python
df_merged = df_merged.dropna()
```

```python
# Round the 'Owned Users' column to 0 decimal places and convert it to integers
df_merged['Owned Users'] = df_merged['Owned Users'].round(0).astype(int)
df_merged['Year Published'] = df_merged['Year Published'].round(0).astype(int)
```

```python
# Calculate the range (min and max)
min_value = df_merged['Owned Users'][df_merged['Owned Users'] > 0].min()
max_value = df_merged['Owned Users'].max()
value_range = max_value - min_value

# Calculate the median
median_value = df_merged['Owned Users'].median()

# Display the results
print(f"Minimum Value: {min_value}")
print(f"Maximum Value: {max_value}")
print(f"Range: {value_range}")
print(f"Median Value: {median_value}")
```

```python
df_merged['Owned_Bool'] = df_merged['Owned Users'] > 311
df_merged.head()
```

```python
df_merged['Owned_Bool'].value_counts()
```

```python
df_merged['Year Offset'] = 2024 - df_merged['Year Published']
```

```python
df_merged['Owned_Per_Year'] = round((df_merged['Owned Users'] / df_merged['Year Offset']),2)
```

```python
# Calculate the range (min and max)
min_value = df_merged['Owned_Per_Year'][df_merged['Owned_Per_Year'] > 0].min()
max_value = df_merged['Owned_Per_Year'].max()
value_range = max_value - min_value

# Calculate the median
median_value = df_merged['Owned_Per_Year'].median()

# Display the results
print(f"Minimum Value: {min_value}")
print(f"Maximum Value: {max_value}")
print(f"Range: {value_range}")
print(f"Median Value: {median_value}")
```

```python
df_merged['perYear_Bool'] = df_merged['Owned_Per_Year'] > 23.33
df_merged.head()
```
