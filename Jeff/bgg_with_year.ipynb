```python
import pandas as pd

# Adjust the delimiter based on your file's structure
df = pd.read_csv('latest_bg.csv', delimiter=',')

# Display the first few rows
print(df.head())
```

```python
import pandas as pd

# Adjust the delimiter based on your file's structure
df2 = pd.read_csv('bgg_dataset.csv', delimiter=';')

# Display the first few rows
print(df2.head())
```

```python
import pandas as pd

# Perform the merge
df_merged = pd.merge(df, df2[['Name', 'Year Published']], 
                     on='Name', 
                     how='left')

df_merged.head()

# Now, df_merged will contain all rows from df_left, with the "year published" column brought in from df_right
```

```python
df_merged.info()
```

```python
null_counts = df_merged.isnull().sum()
print(null_counts)
```

```python
df_merged = df_merged.dropna()
```

```python
# Round the 'Owned Users' column to 0 decimal places and convert it to integers
df_merged['Owned Users'] = df_merged['Owned Users'].round(0).astype(int)
df_merged['Year Published'] = df_merged['Year Published'].round(0).astype(int)
```

```python
df_merged.head()
```

```python
df_merged['Year Offset'] = 2024 - df_merged['Year Published']
```

```python
df_merged['Owned_Per_Year'] = round((df_merged['Owned Users'] / df_merged['Year Offset']),2)
```

```python
df_merged.info()
```

```python
df_merged.to_csv('this_is_it.csv', index=False)
```

```python
df_merged = df_merged.drop(columns=['Bayesian Average', 'bayes_above_average', 'Name', 'Rating Average', 'Year Offset', 'Owned_Per_Year'])
df_merged.head()
```

```python
df_merged = pd.get_dummies(df_merged, columns=['Most Notable Genre', 'Most Notable Mechanic', 'Domain1'])
df_merged.tail()
```

```python
X = df_merged.copy()
X.drop('above_average', axis=1, inplace=True)
X.head()
```

```python
y = df_merged['above_average']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

rf_model = RandomForestClassifier(n_estimators=500, random_state=78)
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
predictions = rf_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
%matplotlib inline

# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Zip feature importances with column names, sort by importance in descending order, and then select top 10
top_10_features = sorted(zip(rf_model.feature_importances_, X.columns), reverse=True)[:10]

# Display the top 10 features
for importance, feature in top_10_features:
    print(f"{feature}: {importance}")
```

```python
# Visualize the top 10 features by importance
importances_df = pd.DataFrame(sorted(zip(rf_model.feature_importances_, X.columns), reverse=True))
importances_df.set_index(importances_df[1], inplace=True)
importances_df.drop(columns=1, inplace=True)
importances_df.rename(columns={0: 'Feature Importances'}, inplace=True)

# Select the top 10 features
top_10_importances = importances_df.head(10)

# Sort and plot the top 10 features
top_10_importances_sorted = top_10_importances.sort_values(by='Feature Importances')
top_10_importances_sorted.plot(kind='barh', color='lightgreen', title='Top 10 Feature Importances', legend=False)
```
